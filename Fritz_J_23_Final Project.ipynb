```python
import numpy as np
from numpy import arange, std, mean, array
from pylab import plot,xlabel,ylabel,show, legend
import statistics

run_alpha = []
run_R = []
run_beta = []
run_Z = []
run_total = []

for run in range(10):
    # Constants
    N_At = 20000
    N_Rn = 0            
    N_Bi215 = 0 
    N_Bi211 = 0
    N_Pb211 = 0
    N_Pb207 = 0
    N_Tl = 0
    tauAt = 56*60  
    tauBi215 = 7.6*60
    tauBi211 = 21.4*60
    tauRn = 4*60
    tauPb211 = 36.1*60
    tauTl = 4.77*60
    h = 1.0                  # Size of time-step in seconds
    pAt = 1 - 2**(-h/tauAt)    # Probability of decay in one step
    pBi215 = 1 - 2**(-h/tauBi215)
    pBi211 = 1 - 2**(-h/tauBi211)
    pRn = 1 - 2**(-h/tauRn)
    pPb211 = 1 - 2**(-h/tauPb211)
    pTl = 1 - 2**(-h/tauTl)
    tmax = 35000               # Total time
    alpha_decay = []
    R_decay = []
    beta_decay = []
    Z_decay = []

# Lists of plot points
    tpoints = arange(0.0,tmax,h)
    Atpoints = [] 
    Rnpoints = [] 
    Bi215points = []
    Bi211points = []
    Pb211points = []
    Pb207points = [] 
    Tlpoints = [] 

# Main loop
    for t in tpoints:    
        Atpoints.append(N_At) # append starting # of At atoms
        Rnpoints.append(N_Rn) # append starting # of Rn atoms
        Bi215points.append(N_Bi215)
        Bi211points.append(N_Bi211)
        Pb211points.append(N_Pb211)
        Pb207points.append(N_Pb207)
        Tlpoints.append(N_Tl)

    # Calculate the number of atoms that decay
        decay1 = 0
        for i in range(N_Tl):
            if np.random.random() < pTl:
                decay1 += 1
                N_Tl -= decay1
                N_Pb207 += decay1
#                beta_decay += 1
                beta_decay.append(1)
    
        decay2 = 0
        for i in range(N_Bi211): # determine number of atoms that decay
            if np.random.random() < pBi211:
                if np.random.random() < 0.997:
                    decay2 += 1
                    N_Bi211 -= 1
                    N_Tl += 1
#                    alpha_decay += 4
                    alpha_decay.append(4)
                else:
                    decay2 += 1
                    N_Bi211 -= 1
                    N_Pb207 += 1
#                    R_decay += 3
                    R_decay.append(3)
               
        decay3 = 0
        for i in range(N_Pb211):
            if np.random.random() < pPb211:
                decay3 += 1
                N_Pb211 -= decay3
                N_Bi211 += decay3
#                beta_decay += 1
                beta_decay.append(1)
            
        decay4 = 0
        for i in range(N_Rn):
            if np.random.random() < pRn:
                decay4 += 1
                N_Rn -= decay4
                N_Pb211 += decay4
#                R_decay += 3
                R_decay.append(3)
            
        decay5 = 0
        for i in range(N_Bi215):
            if np.random.random() < pBi215:
                decay5 += 1
                N_Bi215 -= decay5
                N_Pb211 += decay5
#                Z_decay += 7
                Z_decay.append(7)
            
        decay6 = 0
        for i in range(N_At): 
            if np.random.random() < pAt:
                if np.random.random() < 0.97:
                    decay6 += 1
                    N_At -= 1
                    N_Bi215 += 1
#                    alpha_decay += 4
                    alpha_decay.append(4)
                else:
                    decay6 += 1
                    N_At -= 1
                    N_Rn += 1  
#                    beta_decay += 1
                    beta_decay.append(1)

        
# Make the graph
    plot(tpoints,Atpoints, c='royalblue')
    plot(tpoints,Rnpoints, c='lightcoral')
    plot(tpoints,Bi215points, c='limegreen')
    plot(tpoints,Bi211points, c='darkturquoise')
    plot(tpoints,Pb211points, c='gold')
    plot(tpoints,Pb207points, c='darkorange')
    plot(tpoints,Tlpoints, c='darkorchid')
    xlabel("Time")
    ylabel("Number of atoms")
    legend(["At219", "Rn219", "Bi215", "Bi211", "Pb211", "Pb207", "Tl207"])
    show()


#2
    total_energy_decay = alpha_decay + R_decay + beta_decay + Z_decay

    print("The total alpha decay energy is " + str(sum(alpha_decay)) + " MeV")
    print("The total R decay energy is " + str(sum(R_decay)) + " MeV")
    print("The total beta decay energy is " + str(sum(beta_decay)) + " MeV")
    print("The total Z decay energy is " + str(sum((Z_decay))) + " MeV")
    print("The total decay energy is " + str(sum(total_energy_decay)) + " MeV")

#average and standerd deviation
#   print("Alpha: AVG = " + str(np.mean(alpha_decay)) + "  STD = "+ str(np.std(alpha_decay)))
#   print("R:     AVG = " + str(np.mean(R_decay)) + "    STD = "+ str(np.std(R_decay)))
#   print("Beta:  AVG = " + str(np.mean(beta_decay)) + "   STD = "+ str(np.std(beta_decay)))
#   print("Z:     AVG = " + str(np.mean(Z_decay)) + "   STD = "+ str(np.std(Z_decay)))
#   print("Total: AVG = " + str(np.mean(total_energy_decay)) + "  STD = "+ str(np.std(total_energy_decay)))

    run_alpha.append(sum(alpha_decay))
    run_R.append(sum(R_decay))
    run_beta.append(sum(beta_decay))
    run_Z.append(sum(Z_decay))
    run_total.append(sum(total_energy_decay))
#end of 10 for loop
print("Alpha: AVG = " + str(np.mean(run_alpha)) + "  STD = "+ str(np.std(run_alpha)))
print("R:     AVG = " + str(np.mean(run_R)) + "    STD = "+ str(np.std(run_R)))
print("Beta:  AVG = " + str(np.mean(run_beta)) + "   STD = "+ str(np.std(run_beta)))
print("Z:     AVG = " + str(np.mean(run_Z)) + "   STD = "+ str(np.std(run_Z)))
print("Total: AVG = " + str(np.mean(run_total)) + "  STD = "+ str(np.std(run_total)))

#shield thickness
alpha_avg = np.mean(run_alpha)
alpha_std = np.std(run_alpha)
shield_thickness = (alpha_avg + 3*alpha_std)/1250
print(str(shield_thickness) + " cm shield would be thick enough to safety block all alpha particle decay energy.")

```
