1. Write a ``Particle`` class that can be used to represent a particle with a mass, a 3-d position, and a 3-d velocity.

2. Write a method that can be used to compute the kinetic energy of the particle

3. Write a method that takes another particle as an argument and finds the distance between the two particles

4. Write a method that given a time interval ``dt`` will update the position of the particle to the new position based on the current position and velocity.

5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np

class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        
    def kinetic_energy(self):
        print(0.5*self.mass*(self.vx**2+self.vy**2+self.vz**2))
               
    def find_separation(self, other):
        print(np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2))

    def position_update(self, dt):
        self.x += self.vx*dt
        self.y += self.vy*dt
        self.z += self.vz*dt
        
class ChargedParticle(Particle):

    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge
    
```

```python
p1 = ChargedParticle(1, 1, 1, 1, 1, 1, 1, 1)
p2 = ChargedParticle(1, 2, 2, 2, 1, 1, 1, -1)
p1.find_separation(p2)
p1.position_update(1)
print(p1.x, p1.y, p1.z)
p1.find_separation(p2)
# write/copy code here
```

```python
import numpy as np

class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        
    def kinetic_energy(self):
        print(0.5*self.mass*(self.vx**2+self.vy**2+self.vz**2))
               
    def find_separation(self, other):
        print(np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2))

    def position_update(self, dt):
        self.x = self.x + self.vx*dt
        self.y = self.y + self.vy*dt
        self.z = self.z + self.vz*dt

class ChargedParticle(Particle):

    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge
    def interact(self, other):
        if np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2) <= 0.25: 
            
            if (self.charge > 0 and other.charge < 0) or  (self.charge < 0 and self.charge > 0):
                #mass = self.mass + other.mass
                print("Interaction")
                print("Particles combine")
                self.vx = 0
                self.vy = 0
                self.vz = 0
                other.vx = 0
                other.vy = 0
                other.vz = 0
                self.charge = 0
                other.charge = 0 
                print("Merge")
            elif (self.charge < 0 and other.charge < 0) or  (self.charge > 0 and self.charge > 0): 
                print("Interaction")
                print("Particles Repel")
                self.vx = -1 * self.vx * (self.charge+other.charge) * (self.mass/other.mass)
                self.vy = -1 * self.vy * (self.charge+other.charge) * (self.mass/other.mass)
                self.vz = - self.vz * (self.charge+other.charge) * (self.mass/other.mass)
                other.vx = - other.vx * -1 * (self.charge+other.charge) * (other.mass/self.mass)
                other.vy = - other.vy * -1 * (self.charge+other.charge) * (other.mass/self.mass)
                other.vz = - other.vz * -1 * (self.charge+other.charge) * (other.mass/self.mass)
        #else:
         #   print ("No Interaction")
        
    
```
